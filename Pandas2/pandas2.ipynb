# Problem 1

```python
def prob1():
    """
    Create 3 visualizations of the crime data set.
    One of the visualizations should be a histogram.
    The visualizations should be clearly labelled and easy to understand.
    """
    #read in file
    crime = pd.read_csv('crime_data.csv',index_col='Year')
    
    #histogram
    crime.plot(kind='hist',y=['Murder'],bins=8,color='red',edgecolor='black')
    plt.xlabel("Number of Times Committed")
    plt.ylabel("Frequency in Number Years")
    
    #scatter plot
    crime.plot(kind='scatter',x='Property',y='Larceny')
    plt.xlabel("Number of Property Crimes")
    plt.ylabel("Number of Larceny Crimes")
    
    #Line plot with all crimes
    crime.plot(y=crime.keys()[3:])
    plt.ylabel("Number Committed")
    #with 6 least committed
    crime.plot(y=['Violent','Vehicle Theft','Aggravated Assault','Robbery','Forcible Rape','Murder'])
    plt.ylabel("Number Committed")
    #with 2 least committed
    crime.plot(y=['Forcible Rape','Murder'])
    plt.ylabel("Number Committed")
```

```python
prob1()
```

# Problem 3

```python
def prob3():
    """
    Use crime_data.csv to display the following distributions.
        1. The distributions of Burglary, Violent, and Vehicle Theft
        2. The distributions of Vehicle Theft against the values of Robbery
    """
    plt.rcParams["figure.figsize"] = (20,10) #make plots big
    
    #read in file
    crime = pd.read_csv('crime_data.csv')
    
    crime.plot(kind='box',y=['Burglary','Violent','Vehicle Theft'])
    plt.ylabel('Number of Crimes')
    plt.show()
    
    crime.plot(kind="hexbin", x="Robbery", y="Vehicle Theft", sharex=False, gridsize=20)
```

```python
prob3()
```

```python
def prob4():
    """
    Create 3 plots that compare variables or universities.
    These plots should answer questions about the data, 
    e.g. what is the distribution of graduation rates or do schools 
    with lower student to faculty ratios have higher tuition costs.
    
    These plots should be easy to understand, 
    have clear variable names, and citations.
    """
    plt.rcParams["figure.figsize"] = (20,10) #make plots big
    
    #read in file
    uni = pd.read_csv('college.csv')
    uni.rename(columns={'perc.alumni':'Percentage Alumni Donate','Outstate':'Out of State Tuition','Personal':'Est. Personal Spending',"Enroll":"# New Students Enrolled"},inplace=True)
    uni.rename(columns={'Apps':'# applications received','Accept':'# applications accepted','Grad.Rate':'Graduation Rate'},inplace=True)

    #plot scatter matrix of these following variables that I think might correlate
    pd.plotting.scatter_matrix(uni[['Percentage Alumni Donate','Out of State Tuition','Est. Personal Spending','# New Students Enrolled']])
    plt.show()
    
    #likewise but with different variables
    pd.plotting.scatter_matrix(uni[['# applications received','# applications accepted','# New Students Enrolled','Graduation Rate']])
    plt.show()
    
    #box plot of stud-fac ratio, #terminal degrees, and top10perc
    uni.plot(kind='box',y=['S.F.Ratio','Terminal','Top10perc'])
    plt.xticks([1,2,3],['Student/Faculty Ratio','Perc of Faculty w/ Terminal Degree','Perc of New Students from top 10% of high school class'])
    plt.ylabel("Percentage")
    
    #write source in as xlabel, which is fine since the xticks are labeled
    plt.xlabel("Source: https://cran.r-project.org/web/packages/ISLR/ISLR.pdf")
    
    
```
